# Exploratory Data Analysis (EDA) - Movie Recommender
This notebook explores the MovieLens dataset to understand ratings distribution and sparsity.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set style
sns.set(style="whitegrid")
```

## 1. Load Data

```python
ratings = pd.read_csv('../data/ratings.csv')
movies = pd.read_csv('../data/movies.csv')
print(f"Ratings shape: {ratings.shape}")
print(f"Movies shape: {movies.shape}")
```

## 2. Ratings Distribution

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='rating', data=ratings, palette='viridis')
plt.title('Distribution of Movie Ratings')
plt.show()
```

## 3. Ratings per User

```python
user_counts = ratings['userId'].value_counts()
plt.figure(figsize=(10, 6))
sns.histplot(user_counts, bins=50, kde=True, color='blue')
plt.title('Number of Ratings per User')
plt.xlabel('Number of Ratings')
plt.show()
```

## 4. Ratings per Movie

```python
movie_counts = ratings['movieId'].value_counts()
plt.figure(figsize=(10, 6))
sns.histplot(movie_counts, bins=50, kde=True, color='red')
plt.title('Number of Ratings per Movie')
plt.xlabel('Number of Ratings')
plt.show()
```
